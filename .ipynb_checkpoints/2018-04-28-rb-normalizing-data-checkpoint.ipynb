```python
# %matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
sns.set(color_codes=True)
l = 400

x = (np.random.rand(l) * np.sin(l)) * np.arange(l)
y = -(np.random.rand(l) + 10)
norm_x = (x - np.mean(x)) / np.std(x)
norm_y = (y - np.mean(y)) / np.std(y)
```

```python
fig, ax = plt.subplots(1,2)
sns.distplot(y, ax=ax[0]);
sns.distplot(norm_y, ax=ax[1]);
fig.show()
```

```python
fig_n, ax_n = plt.subplots(1,2)
sns.jointplot(x,y, kind='kde', ax_n=ax_n[0])
sns.jointplot(norm_x, norm_y, kind='kde', ax_n=ax_n[1])
fig_n.show()
```
